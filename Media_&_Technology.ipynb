# Data Preprocessing

```python
# Check for missing values
missing_values = data.isnull().sum()

# Display the missing data statistics
print("Missing Values in Each Column:\n", missing_values)
```

```python
# Remove rows with missing values
data_cleaned = data.dropna()

# Verify that missing values have been removed
print("\nMissing values in each column after removal:")
print(data_cleaned.isnull().sum())
```

```python
# Dropping duplicates
data_cleaned = data_cleaned.drop_duplicates()
```

## 1. Top 10 Youtube Channels based on the number of subscribers

```python
# Sort the DataFrame by 'Subscribers' in descending order
data_sorted = data_cleaned.sort_values(by="subscribers", ascending=False)

# Select and display the top 10 channels
df = data_sorted.head(10)[["Youtuber", "subscribers"]]
df
```

```python
plt.figure(figsize=(20, 10))
plt.bar(df['Youtuber'], df['subscribers'])
plt.xlabel('Youtuber')
plt.ylabel('Subscribers')
plt.title('Top 10 Youtube Channels based on Subscribers')
plt.show()
```

## 2. Which category has the highest average number of subscribers?

```python
# Group by 'category' and calculate the average number of subscribers
category_avg_subs = data_cleaned.groupby('category')['subscribers'].mean()

# Find the category with the highest average number of subscribers
top_category = category_avg_subs.idxmax()
top_avg_subs = category_avg_subs.max()

print(f"The category with the highest average number of subscribers is '{top_category}' with an average of {top_avg_subs:.2f} subscribers.")
```

## 11. Is there a relationship between gross tertiary education enrollment and the number of YouTube channels in a country?

```python
# Group by 'Country' and count the number of channels
country_channel_count = data_cleaned.groupby('Country of origin')['Youtuber'].count().reset_index()
country_channel_count.columns = ['Country of origin', 'YouTube_Channels']

# Merge with the original DataFrame to get the gross tertiary education enrollment
merged_df = pd.merge(country_channel_count, data_cleaned[['Country of origin', 'Gross tertiary education enrollment (%)']].drop_duplicates(), on='Country of origin')

# Calculate the correlation
correlation = merged_df['Gross tertiary education enrollment (%)'].corr(merged_df['YouTube_Channels'])

print(f"Correlation between gross tertiary education enrollment and the number of YouTube channels: {correlation:.2f}")

# Plot the relationship
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Gross tertiary education enrollment (%)'], merged_df['YouTube_Channels'])
plt.xlabel('Gross Tertiary Education Enrollment (%)')
plt.ylabel('Number of YouTube Channels')
plt.title('Relationship between Gross Tertiary Education Enrollment and Number of YouTube Channels by Country')
plt.show()
```

The correlation value is 0.04 which indicates that the two factors are independent of each other.

## 12. How does the unemployment rate vary among the top 10 countries with the highest number of YouTube channels?

```python
# Group by 'Country' and count the number of YouTube channels
country_channel_count = data_cleaned.groupby('Country of origin')['Youtuber'].count().reset_index()

# Merge with the original DataFrame to get the unemployment rate
merged_df = pd.merge(country_channel_count, data_cleaned[['Country of origin', 'Unemployment rate']].drop_duplicates(), on='Country of origin')

# Sort by the number of YouTube channels to get the top 10 countries
top_10_countries = merged_df.sort_values(by='Youtuber', ascending=False).head(10)

# Display the top 10 countries with their unemployment rates
top_10_countries
```

## 13. Average urban population percentage in countries with YouTube channels

```python
# Group by 'Country' and count the number of channels
country_channel_count = data_cleaned.groupby('Country of origin')['Youtuber'].count().reset_index()
country_channel_count.columns = ['Country of origin', 'YouTube_Channels']

# Merge with the original DataFrame to get the urban and total population
merged_df = pd.merge(country_channel_count, data_cleaned[['Country of origin', 'Urban_population', 'Population']].drop_duplicates(), on='Country of origin')

# Calculate the urban population percentage for each country
merged_df['Urban_Population_Percentage'] = (merged_df['Urban_population'] / merged_df['Population']) * 100

# Calculate the average urban population percentage for each country
average_urban_population_percentage = merged_df.groupby('Country of origin')['Urban_Population_Percentage'].mean().reset_index()

print("Average urban population percentage for each country with YouTube channels:")
average_urban_population_percentage
```

## 19. Seasonal trend of number of videos uploaded by Youtube channels based on creation month

```python
# Define the order of months
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Group by creation month and sum the number of videos uploaded
monthly_uploads = data_cleaned.groupby('created_month')['uploads'].sum().reindex(month_order)

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_uploads.index, y=monthly_uploads.values, order=month_order)
plt.title('Total Videos Uploaded by Channel Creation Month')
plt.xlabel('Creation Month')
plt.ylabel('Number of Videos Uploaded')
plt.show()
```

## 20. Average number of subscribers gained per month since the creation of YouTube channels till now

```python
import datetime

# Convert month abbreviations to month numbers
data_cleaned['Creation_Month_Number'] = data_cleaned['created_month'].apply(lambda x: datetime.datetime.strptime(x, '%b').month)

# Extract the current year and month
current_year = pd.Timestamp.now().year
current_month = pd.Timestamp.now().month

# Calculate the number of months each channel has been active
data_cleaned['Months_Active'] = (float(current_year) - data_cleaned['created_year']) * 12 + (float(current_month) - data_cleaned['Creation_Month_Number'])

# Calculate the average number of subscribers gained per month
data_cleaned['Avg_Subscribers_Per_Month'] = data_cleaned['subscribers'] / data_cleaned['Months_Active']

# Display the DataFrame
data_cleaned[['created_year', 'created_month', 'subscribers', 'Months_Active', 'Avg_Subscribers_Per_Month']]
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Avg_Subscribers_Per_Month'], color='forestgreen', kde=True)
plt.title('Distribution of Average Subscribers Per Month')
plt.xlabel('Average Subscribers Per Month')
plt.ylabel('Frequency')
plt.show()
```
